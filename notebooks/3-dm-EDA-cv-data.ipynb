```python

total_nan_rows = df.isna().any(axis=1).sum()

print(f"Total rows with at least one NaN value: {total_nan_rows}")
```

```python
columns_to_analyze = ['work', 'education', 'skills_projects']

for col in columns_to_analyze:
    nan_counts = df.groupby('category')[col].apply(lambda x: x.isna().sum()).reset_index()
    nan_counts.columns = ['category', 'nan_count']
    
    # Plot a horizontal bar chart
    plt.figure(figsize=(8, 6))
    sns.barplot(data=nan_counts, x='nan_count', y='category', palette='viridis', orient='h')
    
    # Add titles and labels
    plt.title(f'NaN Count by Category in {col}', fontsize=16)
    plt.xlabel('NaN Count', fontsize=14)
    plt.ylabel('Category', fontsize=14)
    plt.tight_layout()
    plt.show()
    
    # Close the figure to avoid overlapping
    plt.close()
```

```python
columns_to_analyze = ['work', 'education', 'skills_projects']

for col in columns_to_analyze:
    # Count words in each entry
    df[col + '_word_count'] = df[col].fillna('').apply(lambda x: len(str(x).split()))

    # Display basic statistics
    print(f"Statistics for '{col}':")
    print(df[col + '_word_count'].describe())
    print("\n")
```
